# Clusterin_song
### This notebook applys clustering songs

```python
import pandas as pd
```

```python
hot100 = pd.read_csv('../data/audio_featured_hot100.csv')
nothot100 = pd.read_csv('../data/audio_featured_not_hot100.csv')
```

```python
hot100['set'] = 'H'
hot100.head()
```

```python
nothot100['set'] = 'N'
nothot100.head()
```

```python
songs = pd.concat([hot100, nothot100], ignore_index=True)
songs
```

```python
songs = songs.drop_duplicates()
songs.shape
```

```python
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
songs = songs.reset_index(drop=True)
songs.to_csv('../data/hot100+not_hot_songs.csv', index=False)

numerical_songs = songs.select_dtypes(include='number')
numerical_songs
```

```python
numerical_songs.describe()
```

```python
#drop duration_ms and time-signature because they are considered to be unnecessary factors to recommend.
# List of column names to drop
columns_to_drop = ['duration_ms', 'time_signature']

# Drop the specified columns from the DataFrame
numerical_songs = numerical_songs.drop(columns_to_drop, axis=1)
numerical_songs
```

```python
#apply StandardScaler

import pickle

scaler = StandardScaler()
scaler.fit(numerical_songs)
numerical_songs_scaled = scaler.transform(numerical_songs)
filename = "../scaler/scaler.pickle" # Path with filename

with open(filename, "wb") as file:
        pickle.dump(scaler,file)

numerical_songs_scaled_df = pd.DataFrame(numerical_songs_scaled, columns = numerical_songs.columns)
display(numerical_songs_scaled_df.head())
```

### K means


```python
K = range(8, 25)

inertia = []
silhouette = []

for k in K:
    print("Training a K-Means model with {} clusters! ".format(k))
    print()
    kmeans = KMeans(n_clusters=k,
                    random_state=1234,
                    verbose=1)
    kmeans.fit(numerical_songs_scaled_df)
    
    filename = "../pickle/kmeans_" + str(k) + ".pickle" # Path with filename
    with open(filename, "wb") as file:
        pickle.dump(kmeans,file)

    inertia.append(kmeans.inertia_)
    silhouette.append(silhouette_score(numerical_songs_scaled_df, kmeans.predict(numerical_songs_scaled_df)))


import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

fig, ax = plt.subplots(1,2,figsize=(16,8))
ax[0].plot(K, inertia, 'bx-')
ax[0].set_xlabel('k')
ax[0].set_ylabel('inertia')
ax[0].set_xticks(np.arange(min(K), max(K)+1, 1.0))
ax[0].set_title('Elbow Method showing the optimal k')
ax[1].plot(K, silhouette, 'bx-')
ax[1].set_xlabel('k')
ax[1].set_ylabel('silhouette score')
ax[1].set_xticks(np.arange(min(K), max(K)+1, 1.0))
ax[1].set_title('Silhouette Method showing the optimal k')
```
